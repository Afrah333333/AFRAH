```python
#k folding cross validation - used to decide which model is best suited for an accurate prediction
# the starter way
from sklearn.datasets import load_digits
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

digits = load_digits()
dir(digits)

def get_score(model,X_train , X_test , y_train ,y_test ):
  model.fit(X_train,y_train)
  return model.score(X_test,y_test)

from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)

scores_logistic = []
scores_svm = []
scores_rf = []

for train_index, test_index in folds.split(digits.data,digits.target):
    X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index], \
                                       digits.target[train_index], digits.target[test_index]
    scores_logistic.append(get_score(LogisticRegression(solver='liblinear',multi_class='ovr'), X_train, X_test, y_train, y_test))
    scores_svm.append(get_score(SVC(gamma='auto'), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(n_estimators=40), X_train, X_test, y_train, y_test))
print('Using logistic:',scores_logistic)

print('Using support vector machine:',scores_svm)

print('using random forest:',scores_rf)



#but an easier way is using cross validation score in sklearn
from sklearn.model_selection import cross_val_score
l=cross_val_score(LogisticRegression(),digits.data,digits.target,cv=3)
r=cross_val_score(RandomForestClassifier(),digits.data,digits.target,cv=3)
s=cross_val_score(SVC(),digits.data,digits.target,cv=3)
print(l)
print(s)
print(r)

```

```python
from ipywidgets import Label
#kmeans clustering for unsupervised training where the targets are not known
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
iris = load_iris()

df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.drop(['sepal length (cm)','sepal width (cm)'],axis='columns',inplace=True)
df.head()

#scaling
scaler = MinMaxScaler()
scaler.fit(df[['petal length (cm)']])
df[['petal length (cm)']] = scaler.transform(df[['petal length (cm)']])

scaler.fit(df[['petal width (cm)']])
df[['petal width (cm)']]= scaler.transform(df[['petal width (cm)']])
df

km = KMeans(n_clusters=2)
yp= km.fit_predict(df)

df['clusters'] = yp

df1 = df[df.clusters==0]
df2 = df[df.clusters==1]
df3 = df[df.clusters==2]
df1

centroids = km.cluster_centers_

plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='red')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='blue')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='green')
plt.scatter(centroids[:,0],centroids[:,1],color='purple',marker='+',label='centroids')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.legend()

#elbow plot
sse = [] #sum squared error
k_range = range(1,10)
for k in k_range:
  km = KMeans(n_clusters=k)
  km.fit(df[['petal length (cm)']],df[['petal width (cm)']])
  sse.append(km.inertia_)

sse
plt.xlabel('k')
plt.ylabel('sse')
plt.plot(k_range,sse)

```

```python
from pandas.core.strings.accessor import isna
#PCA- Principal Component Analysis is a dimensionality reduction techinque used in improving performance of your model by generating most relevant features that are nedded to generate an accurate prediction
df = pd.read_csv('heart.csv')
df.head()
def remove_outliers_zscore(df, threshold=3):
    # Calculate the Z-scores for each column
    z_scores_df = (df - df.mean()) / df.std()

    # Identify rows with any Z-score exceeding the threshold
    outlier_indices = z_scores_df.abs() > threshold

    # Remove rows with outliers
    cleaned_df = df[~outlier_indices.any(axis=1)]

    return cleaned_df



# Specify the columns on which to remove outliers (you can choose the relevant columns for your dataset)
columns_to_remove_outliers = ['RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak']

# Remove outliers from the selected columns
df_cleaned = remove_outliers_zscore(df[columns_to_remove_outliers])


df['RestingBP_new'] = df_cleaned['RestingBP']
df['Cholesterol_new'] = df_cleaned['Cholesterol']
df['FastingBS_new'] = df_cleaned['FastingBS']
df['MaxHR_new'] = df_cleaned['MaxHR']
df['Oldpeak_new'] = df_cleaned['Oldpeak']

ds = df.drop( ['RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak'],axis='columns')
ds.columns[ds.isna().any()]

ds.RestingBP_new = ds['RestingBP_new'].fillna(ds['RestingBP_new'].mean())
ds.Cholesterol_new = ds['Cholesterol_new'].fillna(ds['Cholesterol_new'].mean())
ds.FastingBS_new = ds['FastingBS_new'].fillna(ds['FastingBS_new'].mean())
ds.MaxHR_new = ds['MaxHR_new'].fillna(ds['MaxHR_new'].mean())
ds.Oldpeak_new = ds['Oldpeak_new'].fillna(ds['Oldpeak_new'].mean())
ds.columns[ds.isna().any()]

ds.Sex = ds['Sex'].map({
    'M':1,
    'F':2
})

ds.ChestPainType= ds['ChestPainType'].map({
    'ATA':1, 'NAP':2, 'ASY':3, 'TA':4
})

ds.RestingECG = ds['RestingECG'].map({
   'Normal':1, 'ST':2, 'LVH':3
})

ds.ExerciseAngina = ds['ExerciseAngina'].map({
    'N':1,
    'Y':2
})

ds.ST_Slope = ds['ST_Slope'].map({
    'Up':1, 'Flat':2, 'Down':3
})
ds
X = ds.drop('HeartDisease', axis='columns')
y = ds['HeartDisease']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

data_model = [svm.SVC(),RandomForestClassifier(),LogisticRegression()]
cv = [3,5,10]
avg_scores={}
for dm in data_model:
  for c in cv:
    scores = cross_val_score( dm, X , y , cv=c)
    model_name = str(dm)
    avg_scores[model_name + '_' + str(c)] = np.average(scores)


avg_scores
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)

model = svm.SVC()
model.fit(X_train,y_train)
model.score(X_test,y_test)

from sklearn.decomposition import PCA
pca = PCA(0.95)
pca_x = pca.fit_transform(X)

Xpca_train, Xpca_test, y_train, y_test = train_test_split(pca_x,y,test_size=0.2,random_state=1)
model = svm.SVC()
model.fit(Xpca_train,y_train)
model.score(Xpca_test,y_test)

#Bagging- method used to control overfitting
#bagging with SVC
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(base_estimator=svm.SVC(), n_estimators=100, max_samples=0.8, random_state=0)
scores = cross_val_score(bag_model, pca_x, y, cv=5)
print('SVC with bagging',scores.mean())
scores = cross_val_score(svm.SVC(), pca_x, y, cv=5)
print('SVC without bagging',scores.mean())
#as you see the results has no big difference meaning we have no high variance
#bagging on decision trees
bag_model2 = BaggingClassifier(base_estimator=DecisionTreeClassifier(random_state=0), n_estimators=100, max_samples=0.8, random_state=0,oob_score=True,)
scores = cross_val_score(bag_model2, pca_x, y, cv=5)
print('DecisionTree with bagging',scores.mean())
scores = cross_val_score(DecisionTreeClassifier(random_state=0), pca_x, y, cv=5)
print('DecisionTree without bagging',scores.mean())
#the difference is wide here meaning that its better to bag while using decisiontreeclassifier as variance is high

#random forest already uses bagging underneath whereby it resamples both rows and columns
from sklearn.ensemble import RandomForestClassifier

scores = cross_val_score(RandomForestClassifier(), pca_x, y, cv=5)
print('RandomForest bagging',scores.mean())


```
